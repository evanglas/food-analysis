```python
import pandas as pd
import json
```

```python
df = pd.read_csv("data/nutrient_constraints.csv")
```

```python
constraints = {}
for i, row in df.iterrows():
    nutrient_nbrs = row["nutrient_nbrs"]
    if nutrient_nbrs not in constraints:
        constraints[nutrient_nbrs] = {}
        constraints[nutrient_nbrs]["nutrient_name"] = row["constraint_name"]
        constraints[nutrient_nbrs]["unit_name"] = row["unit_name"]
        constraints[nutrient_nbrs]["constraints"] = {}
    constraints[nutrient_nbrs]["constraints"][row["constraint_type"]] = row[
        "constraint_value"
    ]
```

```python
with open("data/nutrient_constraints_2.json", "w") as f:
    json.dump(constraints, f)
```

```python
nutrient_infos = {}
for i, row in df_single.iterrows():
    if row["nutrient_nbrs"] not in nutrient_infos:
        nutrient_info = {}
        nutrient_info["nutrient_name"] = row["constraint_name"]
        nutrient_info["nutrient_nbr"] = int(row["nutrient_nbrs"])
        nutrient_info["unit_name"] = row["unit_name"]
        nutrient_info["nutrient_rdas"] = {}
        nutrient_info["nutrient_rdas"]["default"] = []
        nutrient_info["nutrient_rdas"]["default"].append(
            {row["constraint_type"]: row["constraint_value"]}
        )
        nutrient_infos[nutrient_info["nutrient_nbr"]] = nutrient_info
    else:
        nutrient_infos[row["nutrient_nbrs"]]["nutrient_rdas"]["default"].append(
            {row["constraint_type"]: row["constraint_value"]}
        )
with open("data/nutrient_info.json", "w") as f:
    json.dump(nutrient_infos, f)
```
