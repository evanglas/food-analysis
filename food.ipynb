```python
import pandas as pd
```

```python
nutrition = pd.read_excel('2019-2020 FNDDS At A Glance - FNDDS Nutrient Values.xlsx', skiprows=1)
```

```python
prices = pd.read_excel('TFP_2021_Online_Supplement.xlsx', sheet_name='Online Supplement Data')
```

```python
nutrition_prices = prices.merge(nutrition, how='inner', left_on='FNDDS Food Code', right_on='Food code')
```
